```python
df['Date of injury resulting in death (month/day/year)'] = pd.to_datetime(df['Date of injury resulting in death (month/day/year)'])
```

```python
df['Location of death (city)'].value_counts()
```

```python
city = df['Location of death (city)'].value_counts(ascending=False)

df_city = df.filter(["Location of death (city)"], axis=1)
df_city["count"] = 1

grouped_city = df_city.groupby("Location of death (city)", as_index=False,sort=False).sum()
grouped_city.sort_index(ascending = False)

grouped_city = grouped_city.sort_values("count", ascending=False).head(5)                                                       

plt.figure(figsize=(10,5))
sns.barplot(data=grouped_city, x="Location of death (city)", y="count")
plt.title("Number of vicitims by city", fontsize=15);
```
